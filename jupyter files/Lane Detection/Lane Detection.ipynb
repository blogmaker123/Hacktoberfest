# LANE DETECTION

```python
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
import numpy as np
import cv2
```

```python
c_image = mpimg.imread('lane.jpg')
plt.imshow(c_image)
c_image.shape
```

```python
g_image = cv2.cvtColor(c_image, cv2.COLOR_BGR2GRAY)
plt.imshow(g_image, cmap = 'gray')
g_image.shape
```

## SORTING WHITE PIXELS

```python
image_copy = np.copy(g_image)
image_copy.shape
```

```python
image_copy[ (image_copy[:,:] < 245) ] = 0

plt.imshow(image_copy, cmap = 'gray')
plt.show()
```

## REGION OF INTEREST MASKING

```python
g_image = cv2.cvtColor(c_image, cv2.COLOR_BGR2GRAY)
plt.imshow(g_image, cmap = 'gray')
g_image.shape
```

```python
height, width = g_image.shape[:2]
```

```python
height
```

```python
width
```

```python
ROI = np.array([[(0, height),(400, 330), (550, 330), (width, height)]], dtype=np.int32)    
blank = np.zeros_like(g_image) 

blank.shape
```

```python
mask = cv2.fillPoly(blank, ROI, 255)

masked_image = cv2.bitwise_and(g_image, mask)
masked_image.shape
```

```python
plt.imshow(masked_image, cmap = 'gray')
```
